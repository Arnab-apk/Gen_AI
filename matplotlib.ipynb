```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
Data={'Year':[1920,1930,1940,1950,1960,1970,1990,2000,2010,2020],
      'Exchange Rate':[69,71,64,62,59,72,71,75,78,81]}
```

```python
type(Data)
```

```python
#by default it can also work without teh column names
df=pd.DataFrame(Data,columns =['Year','Exchange Rate'])
```

```python
df
```

```python
type(df)
```

```python
#Bar chart
df.plot(x='Year',y='Exchange Rate',kind='bar')
plt.show()
```

```python
#Line chart
df.plot(x='Year',y='Exchange Rate',kind="line")
plt.show()
```

```python
df.plot(x='Year', y='Exchange Rate', kind='area')
plt.show()
```

```python
df.plot(x='Year', y='Exchange Rate', kind='barh')
plt.show()
```

```python
df.plot(x='Year', y='Exchange Rate', kind='scatter')
plt.show()
```
