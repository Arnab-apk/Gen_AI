# Special Character Removal
Removing non-alphanumeric characters keeps text clean for downstream NLP tasks such as tokenization and matching.
- Reduce noise (punctuation/symbols) before analysis
- Improve text quality for tokenization and vectorization
- Keep spaces so word boundaries remain intact


```python
import re
#input string
input_str="hello how are you@"
#using regular expression to remove special characters
clean_str=re.sub(r"[^a-zA-Z0-9\s]","",input_str)
print(clean_str)
```

## Libraries
- `re`: Python regex for quick cleaning (used here)
- `spaCy`: Full-featured NLP pipeline (tokenization, POS, NER)
- `nltk`: Classic NLP toolkit with corpora and preprocessors


### Takeaway
Use a simple regex (`re.sub(r"[^a-zA-Z0-9\s]", "", text)`) when you just need to strip symbols while preserving word boundaries. For richer preprocessing, hand off the cleaned text to spaCy or NLTK.
