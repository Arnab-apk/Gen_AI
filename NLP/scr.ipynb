__SPECIAL CHARACTER REMOVAL__<br>
__eliminating non alpha numeric characters in the text__<br>
__Use of it__<br>
__noise reduction__<br>
__improved text quality__<br>
__facilitating tokenisation__<br>

```python
import re
#input string
input_str="hello how are you@"
#using regular expression to remove special characters
clean_str=re.sub(r"[^a-zA-Z0-9\s]","",input_str)
print(clean_str)
```
